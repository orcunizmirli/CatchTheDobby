```python
image1 = face_recognition.load_image_file('608fda712269a22e102f6515.jpg')
image2 = face_recognition.load_image_file('elon-musk-reuters_16_9_1594283053-880x495-1_16_9_1620392012-880x495.jpg')
image3 = face_recognition.load_image_file('54ed86d061361f15b0bd42f9.jpg')
image4 = face_recognition.load_image_file('Steve-jobs-hakkinda-11-ilginc-bilgi-696x464.jpg')
```

```python
face_locations = face_recognition.face_locations(image1)
face_locations2 = face_recognition.face_locations(image2)
face_locations3 = face_recognition.face_locations(image3)
face_locations4 = face_recognition.face_locations(image4)
```

```python
e1 = face_recognition.face_encodings(image1)[0]
e2 = face_recognition.face_encodings(image2)[0]
e3 = face_recognition.face_encodings(image3)[0]
e4 = face_recognition.face_encodings(image4)[0]
```

```python
(top, right, bottom, left) = face_locations[0]
(top2, right2, bottom2, left2) = face_locations2[0]
(top3, right3, bottom3, left3) = face_locations3[0]
(top4, right4, bottom4, left4) = face_locations4[0]
```

```python
view_image = Image.fromarray(image1)

draw = ImageDraw.Draw(view_image)

draw.rectangle(((left, top), (right, bottom)), outline=(0, 255, 255))

view_image
```

```python
view_image = Image.fromarray(image2)

draw = ImageDraw.Draw(view_image)

draw.rectangle(((left2, top2), (right2, bottom2)), outline=(0, 255, 255))

view_image
```

```python
view_image = Image.fromarray(image3)

draw = ImageDraw.Draw(view_image)

draw.rectangle(((left3, top3), (right3, bottom3)), outline=(0, 255, 255))

view_image
```

```python
view_image = Image.fromarray(image4)

draw = ImageDraw.Draw(view_image)

draw.rectangle(((left4, top4), (right4, bottom4)), outline=(0, 255, 255))

view_image
```

```python
result1 = face_recognition.compare_faces([e1], e2)

result2 = face_recognition.compare_faces([e3], e4)

result3 = face_recognition.compare_faces([e3],e1)

print(result1)
print(result2)
print(result3)
```
